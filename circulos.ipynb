```python
%pylab inline
from scipy import fftpack
from scipy.fftpack import fft, fftfreq
data = loadtxt("circulos.dat")
s = 50**-1
x=data[:,0]+3
y=data[:,1]+3
scatter(x,y,s=s)
```

```python
h=250
d=6.0/float(h)
particion=zeros((h,h))
for i in range (0,h):
    for j in range (0,h):
        a=len(x[(x<=d*float(i+1))&(x>d*float(i)) &(y<=d*float(j+1))&(y>d*float(j))])
        particion[i][j]=a
            
    
    
    
```

```python
fig=figure(1,figsize = (10,10))
imshow(particion,cmap=cm.Greys_r)
```

```python
im_ft = fftpack.fft2(particion)
im_ft_shift = fftpack.fftshift(im_ft)
psd2D= abs(im_ft_shift)**2
fig = figure(1, figsize=(10,10))
imshow(log(psd2D[:,:]),cmap=cm.Greys_r)
```
